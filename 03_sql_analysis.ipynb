PHASE 3 — SQL BUSINESS ANALYSIS

```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv("../data/cleaned/superstore_final.csv")

df.head()
```

CREATE SQL DATABASE (IN MEMORY)

```python
conn = sqlite3.connect(":memory:")
```

LOAD DATA INTO SQL TABLE

```python
df.to_sql("sales", conn, index=False, if_exists="replace")
```

```python
pd.read_sql("SELECT * FROM sales LIMIT 5", conn)
```

SQL QUERIES

QUERY 1 — Total Sales & Profit

```python
pd.read_sql("""
SELECT 
    SUM(Sales) AS Total_Sales,
    SUM(Profit) AS Total_Profit
FROM sales
""", conn)
```

QUERY 2 — Sales by Category

```python
pd.read_sql("""
SELECT Category,
       SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Category
ORDER BY Total_Sales DESC
""", conn)
```

QUERY 3 — Profit by Category

```python
pd.read_sql("""
SELECT Category,
       SUM(Profit) AS Total_Profit
FROM sales
GROUP BY Category
ORDER BY Total_Profit DESC
""", conn)
```

QUERY 4 — Sales by Region

```python
pd.read_sql("""
SELECT Region,
       SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Region
ORDER BY Total_Sales DESC
""", conn)
```

QUERY 5 — Top Customers

```python
pd.read_sql("""
SELECT "Customer Name",
       SUM(Sales) AS Total_Sales
FROM sales
GROUP BY "Customer Name"
ORDER BY Total_Sales DESC
LIMIT 10
""", conn)
```

QUERY 6 — Top Products

```python
pd.read_sql("""
SELECT "Product Name",
       SUM(Sales) AS Total_Sales
FROM sales
GROUP BY "Product Name"
ORDER BY Total_Sales DESC
LIMIT 10
""", conn)
```

QUERY 7 — Loss Making Products

```python
pd.read_sql("""
SELECT "Product Name",
       SUM(Profit) AS Total_Profit
FROM sales
GROUP BY "Product Name"
HAVING Total_Profit < 0
ORDER BY Total_Profit ASC
""", conn)
```

QUERY 8 — Monthly Sales Trend

```python
pd.read_sql("""
SELECT "Order Month",
       SUM(Sales) AS Monthly_Sales
FROM sales
GROUP BY "Order Month"
""", conn)
```

QUERY 9 — Sales by Segment

```python
pd.read_sql("""
SELECT Segment,
       SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Segment
""", conn)
```

QUERY 10 — Shipping Performance

```python
pd.read_sql("""
SELECT "Ship Mode",
       AVG("Shipping Days") AS Avg_Shipping_Days
FROM sales
GROUP BY "Ship Mode"
""", conn)
```

QUERY 11 — Average Order Value

```python
pd.read_sql("""
SELECT AVG(Sales) AS Avg_Order_Value
FROM sales
""", conn)
```

QUERY 12 — High Discount Impact

```python
pd.read_sql("""
SELECT Discount,
       AVG(Profit) AS Avg_Profit
FROM sales
GROUP BY Discount
ORDER BY Discount
""", conn)
```

QUERY 13 — Category + Region Analysis

```python
pd.read_sql("""
SELECT Category, Region,
       SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Category, Region
ORDER BY Total_Sales DESC
""", conn)
```

QUERY 14 — Best Customers by Profit

```python
pd.read_sql("""
SELECT "Customer Name",
       SUM(Profit) AS Total_Profit
FROM sales
GROUP BY "Customer Name"
ORDER BY Total_Profit DESC
LIMIT 10
""", conn)
```

QUERY 15 — Profit Margin Analysis

```python
pd.read_sql("""
SELECT Category,
       AVG("Profit Margin") AS Avg_Profit_Margin
FROM sales
GROUP BY Category
""", conn)
```
